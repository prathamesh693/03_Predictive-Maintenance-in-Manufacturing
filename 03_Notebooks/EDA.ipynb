###  **Import Required Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### **Load Dataset**

```python
# Load original dataset
df = pd.read_csv("R:/Projects/1_Data_Science & ML_Projects/03_Predictive Maintenance in Manufacturing/02_Data/predictive_maintenance.csv")

predicted_df=pd.read_csv("R:/Projects/1_Data_Science & ML_Projects/03_Predictive Maintenance in Manufacturing/06_Outputs/prediction_result.csv")
```

### **Initial Data Checks** 

```python
# First 5 data set
print(df.head())
```

```python
# Information about columns.
print(df.info())
```

```python
# Statistical Summary & Null values checking
print(df.describe())
print(df.isnull().sum())  # Check for missing values
```

### **1. Scatter Plot**

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Air temperature [K]', y='Process temperature [K]', hue='Target')
plt.title('Scatter Plot: Air vs Process Temperature')
plt.show()
```

### **2. Line Chart**

```python
plt.figure(figsize=(10, 5))
plt.plot(df['Rotational speed [rpm]'][:200], label='Rotational Speed')
plt.title('Line Chart: Rotational Speed')
plt.xlabel('Sample')
plt.ylabel('RPM')
plt.legend()
plt.show()
```

### **3. Box Plot**

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Failure Type', y='Torque [Nm]', data=df)
plt.title('Box Plot: Torque by Failure Type')
plt.xticks(rotation=45)
plt.show()
```

### **4. Bar Chart**

```python
plt.figure(figsize=(8, 6))
df['Failure Type'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Bar Chart: Failure Type Count')
plt.xlabel('Failure Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

### **5. Pie Chart**

```python
plt.figure(figsize=(6, 6))
df['Type'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart: Machine Type Distribution')
plt.ylabel('')
plt.show()
```

### **6. Histogram**

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Tool wear [min]'], bins=30, kde=True)
plt.title('Histogram: Tool Wear')
plt.xlabel('Tool Wear [min]')
plt.show()
```

### **7. Heatmap**

```python
plt.figure(figsize=(12, 8))
corr = df.select_dtypes(include='number').corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

### **8. Pairplot**

```python
sns.pairplot(df[['Air temperature [K]', 'Process temperature [K]', 'Rotational speed [rpm]',
                 'Torque [Nm]', 'Tool wear [min]', 'Target']], hue='Target')
plt.suptitle('Pairplot: Key Features by Target', y=1.02)
plt.show()
```

### **9. Actual Vs Predicted**

```python
# Plot for RandomForest predictions
sns.countplot(x='Prediction (RandomForest)', data=predicted_df)
plt.title('Prediction Count - Random Forest')
plt.xlabel('Predicted Class')
plt.ylabel('Count')
plt.show()
```

```python
# Plot for XGBoost predictions
sns.countplot(x='Prediction (XGBoost)', data=predicted_df)
plt.title('Prediction Count - XGBoost')
plt.xlabel('Predicted Class')
plt.ylabel('Count')
plt.show()
```
